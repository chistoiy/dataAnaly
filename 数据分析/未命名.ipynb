```python
#按列对比合并：
#01
import pandas as pd 
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
print(left)
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
print(right)
```

```python
result = pd.merge(left, right,how='outer', on=['key1', 'key2'])
#分别根据两个数组的指定列进行对比，如果两个列同时相同的行取出来形成新的数组，如果其中一个数组有多行满足，则另一个数组同一行被多次组合
#how='inner','left','right','outer',不设置how，则默认为inner，right则为根据右边的数组尽可能从左边寻找，找不到的则填充nan，outer则是全部都合并在一起
print(result)
```

```python
#02
#如果两个数组没有相同列名，则使用left_on与right_on指定对比列
result = pd.merge(left, right,how='inner', left_on='key1',right_on='key2')
print(result)
```
